# 4. Longest match recorded

```python
df_recorded_match = (df_match.groupby(['match_id'], as_index=False)[['start_datetime', 'current_period_start_timestamp']]
                    .agg(['min', 'max']))

# Removed null values according to 'current_period_start_timestamp' column 
df_recorded_match = df_recorded_match.loc[(df_recorded_match.loc[:, ('current_period_start_timestamp', 'min')]
                                        .dropna()
                                        .index), :].reset_index(drop=True)
df_recorded_match.head() 
```

```python
diff_min_min = np.diff(df_recorded_match.loc[:, [('start_datetime', 'min'), ('current_period_start_timestamp', 'min')]]).argmax()
diff_min_max = np.diff(df_recorded_match.loc[:, [('start_datetime', 'min'), ('current_period_start_timestamp', 'max')]]).argmax()
diff_max_max = np.diff(df_recorded_match.loc[:, [('start_datetime', 'max'), ('current_period_start_timestamp', 'max')]]).argmax()
diff_max_min = np.diff(df_recorded_match.loc[:, [('start_datetime', 'max'), ('current_period_start_timestamp', 'min')]]).argmax()

print(f"diff min to min col: {diff_min_min}\
        \ndiff min to max col: {diff_min_max}\
        \ndiff max to max col: {diff_max_max}\
        \ndiff max to min col: {diff_max_min}")
```

```python
print(f"Record time for index 5084:\
      \n{df_recorded_match.loc[5084, [('start_datetime', 'min'), ('current_period_start_timestamp', 'min')]]}\
      \n\nRecord time for index 392:\
      \n{df_recorded_match.loc[392, [('start_datetime', 'max'), ('current_period_start_timestamp', 'max')]]}")
```

```python
print(f"Diffrence time for both indexes 5084 and 392 is:\
      \n{pd.Timestamp(1.716569e+09, unit='s') - pd.Timestamp(1.716378e+09, unit='s')}\
      \n{pd.Timestamp(1.714740e+09, unit='s') - pd.Timestamp(1.714645e+09, unit='s') }")

print(f"\nlongest match recorded is:\
 0{pd.Timestamp(1.716569e+09, unit='s').time().hour - pd.Timestamp(1.716378e+09, unit='s').time().hour}\
:0{pd.Timestamp(1.716569e+09, unit='s').time().minute - pd.Timestamp(1.716378e+09, unit='s').time().minute}\
:{pd.Timestamp(1.716569e+09, unit='s').time().second - pd.Timestamp(1.716378e+09, unit='s').time().second} by Match id: 12355549")
```

# 7. Average number of aces per match
### when a player wins a point directly from their serve.

```python
total_aces = (df_statistics[(df_statistics['statistic_category_name'] == 'service') & (df_statistics['statistic_name'] == 'aces')]
              .value_counts())

(df_match
.merge(df_statistics[df_statistics['statistic_name'] == 'aces'], left_on='match_id', right_on='match_id')[['match_id', 'statistic_name']]
.groupby(by=['match_id', 'statistic_name'], as_index=False)['statistic_name']
.agg({'count_aces': lambda x: x.count(), 'average_aces': lambda x: x.count()/total_aces})
.head())
```

# 11. Average duration of matches

```python
diff_min_min = np.diff(df_recorded_match.loc[:, [('start_datetime', 'min'), ('current_period_start_timestamp', 'min')]]).mean()
diff_min_max = np.diff(df_recorded_match.loc[:, [('start_datetime', 'min'), ('current_period_start_timestamp', 'max')]]).mean()
diff_max_max = np.diff(df_recorded_match.loc[:, [('start_datetime', 'max'), ('current_period_start_timestamp', 'max')]]).mean()
diff_max_min = np.diff(df_recorded_match.loc[:, [('start_datetime', 'max'), ('current_period_start_timestamp', 'min')]]).mean()

print(f"diff min to min col: {diff_min_min}\
        \ndiff min to max col: {diff_min_max}\
        \ndiff max to max col: {diff_max_max}\
        \ndiff max to min col: {diff_max_min}")
```

```python
mean_time = (pd.Timestamp((diff_min_min + 
              diff_min_max +
              diff_max_max +
              diff_max_min) / 4, unit='s')
                                        .time())

print(f"Average duration of matches is: {mean_time}")
```

# 16. A player has the highest winning

```python
total_wining = (df_match
.merge(df_odds, left_on='match_id', right_on='match_id')[['player_id', 'current_rank','winnig']]
.sort_values('current_rank', ascending=False)
.head(10)['winnig']
.sum())

(df_match
.merge(df_odds, left_on='match_id', right_on='match_id')[['player_id', 'current_rank','winnig']]
.sort_values('current_rank', ascending=False)
.head(10)
.groupby(by=['player_id'], as_index=False)['winnig']
.agg({'percent_winnig': lambda x : x.sum() * 100/ total_wining}))
```
