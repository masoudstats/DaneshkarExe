```python
import pandas as pd
import os 
import zipfile
import pyarrow.parquet as pq
import warnings
```

```python
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
zip_file_path = "E:\\Daneshkar\\part4\\exe3\\data\\raw\\"
zip_name = [filename for filename in os.listdir(zip_file_path)]

for  name in  zip_name:
    exract_path = zip_file_path + name[:-4] 

    os.makedirs(exract_path, exist_ok=True)

    with zipfile.ZipFile(zip_file_path + name , 'r') as zip_ref:
        zip_ref.extractall(exract_path)
```

```python
total_path = os.listdir(r'E:/Daneshkar/part4/exe3/data/raw/')
total_path_folders_files = [r'E:/Daneshkar/part4/exe3/data/raw/' + total_path[i] for i in range(len(total_path))]
```

```python
match_dataframe = {}
```

```python
i = 0
for idx, file_dir in enumerate(total_path_folders_files):
    if os.path.isdir(file_dir):
        i += 1
        match_dataframe[i] = pd.DataFrame()
        folder_name = file_dir.split('/')[-1]
        print(f"You are in: {file_dir}")
        for sub_folder in os.listdir(file_dir + r'/data/raw'):
            if sub_folder.startswith('raw_match'):
                path = file_dir + '/' + 'data/raw/' + sub_folder
                files = os.listdir(path)
                for file in files:
                    single_datafram = pd.read_parquet(path + '/'+ file)
                    match_dataframe[i] = pd.concat([ match_dataframe[i], single_datafram], axis=0)
```

```python
match_tb = pd.concat(match_dataframe.values(), ignore_index=True)
match_tb.head()
```
