```python
sns.set_style("darkgrid")

ax = sns.barplot(data=df_match['plays'].value_counts(normalize=True), palette='flare')
ax.bar_label(ax.containers[0], fontsize=8)
ax.bar_label(ax.containers[1], fontsize=8)
ax.bar_label(ax.containers[2], fontsize=8)
plt.xlabel('Type of players')

fig = ax.get_figure()
fig.savefig(r'E:\Daneshkar\part4\exe3\reports\figures\Q13_bar.png')  
plt.show()
```
